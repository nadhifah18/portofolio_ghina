## **Menggabungkan Data**

```python
import pandas as pd

# Mengonversi setiap dataset ke format panjang (long format)
ika_long = data1.melt(id_vars=['Provinsi'], var_name='Tahun', value_name='ika')
iku_long = data2.melt(id_vars=['Provinsi'], var_name='Tahun', value_name='iku')
ikl_long = data3.melt(id_vars=['Provinsi'], var_name='Tahun', value_name='ikl')

# Menambahkan data nol pada data ikal untuk tahun 2011-2018
tahun_ikal_tambahan = pd.DataFrame({
    'Provinsi': data4['Provinsi'].repeat(8).reset_index(drop=True),
    'Tahun': [str(y) for y in range(2011, 2019)] * len(data4),
    'ikal': [0] * 8 * len(data4)
})
ikal_long = data4.melt(id_vars=['Provinsi'], var_name='Tahun', value_name='ikal')
ikal_long = pd.concat([tahun_ikal_tambahan, ikal_long], ignore_index=True)

# Mengonversi data IKLH ke format panjang (long format)
iklh_long = data5.melt(id_vars=['Provinsi'], var_name='Tahun', value_name='iklh')

# Menggabungkan seluruh dataset
merged_data = pd.merge(ika_long, iku_long, on=['Provinsi', 'Tahun'], how='outer')
merged_data = pd.merge(merged_data, ikl_long, on=['Provinsi', 'Tahun'], how='outer')
merged_data = pd.merge(merged_data, ikal_long, on=['Provinsi', 'Tahun'], how='outer')
merged_data = pd.merge(merged_data, iklh_long, on=['Provinsi', 'Tahun'], how='outer')

# Mengatur format menjadi 2 angka di belakang koma
merged_data[['ika', 'iku', 'ikl', 'ikal', 'iklh']] = merged_data[['ika', 'iku', 'ikl', 'ikal', 'iklh']].round(2)

# Mengurutkan berdasarkan Provinsi dan Tahun
merged_data = merged_data.sort_values(by=['Provinsi', 'Tahun']).reset_index(drop=True)

# Tampilkan hasil gabungan
print(merged_data)
```

```python
# Menyimpan hasil gabungan ke dalam file CSV
merged_data.to_csv('data_iklh.csv', index=False)
```

### **Spliting Data**

```python
# Pilih hanya kolom ika, iku, ikl, ikal untuk X dan kolom iklh untuk y
X = df[['Tahun', 'ika', 'iku', 'ikl', 'ikal']]
y = df['iklh']

# Pisahkan data berdasarkan tahun
X_train = X[X['Tahun'] < 2023].drop(columns=['Tahun'])
y_train = y[X['Tahun'] < 2023]

X_valid = X[X['Tahun'] == 2023].drop(columns=['Tahun'])
y_valid = y[X['Tahun'] == 2023]

# Menampilkan jumlah sampel dalam training set dan validation set
print(f'Training set: {X_train.shape[0]} samples')
print(f'Validation set: {X_valid.shape[0]} samples')
```

### **1. Linear Regression**

```python
model_linreg = LinearRegression()
model_linreg.fit(X_train, y_train)
y_pred_linreg = model_linreg.predict(X_valid)
```

```python
# Dapatkan nilai koefisien (slope) dan intercept
slope = model_linreg.coef_
intercept = model_linreg.intercept_

# Tampilkan hasil
print("Slope (Koefisien) : ", slope)
print("Intercept (Titik Potong) : ", intercept)
```

```python
import pandas as pd
from IPython.display import display

# Pastikan indeks X_valid sesuai dengan indeks data provinsi
hasil_prediksi = pd.DataFrame({
    'Provinsi': df.loc[X_valid.index, 'Provinsi'],  # Mengambil nama provinsi berdasarkan indeks
    'Actual': y_valid.values,
    'Prediction': y_pred_linreg,
    'Error': y_valid.values - y_pred_linreg,
    'Absolute Error': abs(y_valid.values - y_pred_linreg)
})

# Menampilkan semua data dalam bentuk tabel
display(hasil_prediksi)
```

```python
import pandas as pd
from IPython.display import display

# Pastikan indeks X_valid sesuai dengan indeks data provinsi
hasil_prediksi = pd.DataFrame({
    'Provinsi': df.loc[X_valid.index, 'Provinsi'],  # Mengambil nama provinsi berdasarkan indeks
    'Actual': y_valid.values,
    'Prediction': y_pred_linreg,
    'Percentage Error': (abs(y_valid.values - y_pred_linreg) / y_valid.values) * 100  # Persentase error
})

# Menghitung Mean Absolute Percentage Error (MAPE)
mape = hasil_prediksi['Percentage Error'].mean()

# Menghitung akurasi model (100% - MAPE)
accuracy = 100 - mape

# Menampilkan hasil dalam bentuk tabel
display(hasil_prediksi)

# Menampilkan MAPE dan akurasi
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
print(f"Akurasi Model: {accuracy:.2f}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error
from IPython.display import display

# Pastikan indeks X_valid sesuai dengan indeks data provinsi
hasil_prediksi = pd.DataFrame({
    'Provinsi': df.loc[X_valid.index, 'Provinsi'],  # Mengambil nama provinsi berdasarkan indeks
    'Actual': y_valid.values,
    'Prediction': y_pred_linreg,
    'Error': y_valid.values - y_pred_linreg,
    'Absolute Error': abs(y_valid.values - y_pred_linreg),
    'MAPE': (abs(y_valid.values - y_pred_linreg) / y_valid.values) * 100  # Persentase error
})

# Menghitung evaluasi metrik
mape = hasil_prediksi['MAPE'].mean()
rmse = np.sqrt(mean_squared_error(y_valid, y_pred_linreg))
r2 = r2_score(y_valid, y_pred_linreg)
hasil_prediksi['RMSE'] = np.sqrt((hasil_prediksi['Error'] ** 2)).round(4)  # RMSE per baris
hasil_prediksi['R²'] = r2  # R² nilai yang sama untuk setiap baris

# Menambahkan kolom Accuracy, RMSE, dan R² per baris
hasil_prediksi['Accuracy (%)'] = (100 - hasil_prediksi['MAPE']).round(2)  # Versi numerik untuk perhitungan
hasil_prediksi['Accuracy'] = hasil_prediksi['Accuracy (%)'].astype(str) + '%'  # Versi string untuk tampilan


# Menampilkan hasil dalam bentuk tabel
display(hasil_prediksi[['Provinsi', 'Actual', 'Prediction','RMSE', 'R²','MAPE', 'Accuracy']])

# Menampilkan evaluasi keseluruhan model
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"Koefisien Determinasi (R²): {r2:.4f}")
print(f"Rata-rata Akurasi Model: {hasil_prediksi['Accuracy (%)'].mean():.2f}%")
```

```python
def kategori_nilai(nilai):
    if nilai >= 90:
        return "Sangat Baik"
    elif 70 <= nilai < 90:
        return "Baik"
    elif 50 <= nilai < 70:
        return "Sedang"
    elif 25 <= nilai < 50:
        return "Buruk"
    else:
        return "Sangat Buruk"

# Tambahkan kategori ke dalam DataFrame
hasil_prediksi['Actual Category'] = hasil_prediksi['Actual'].apply(kategori_nilai)
hasil_prediksi['Prediction Category'] = hasil_prediksi['Prediction'].apply(kategori_nilai)

# Tampilkan hasil
display(hasil_prediksi)
```

```python
# Tambahkan kolom perbedaan kategori
hasil_prediksi['Category Match'] = hasil_prediksi['Actual Category'] == hasil_prediksi['Prediction Category']

# Tampilkan hanya baris yang kategorinya berbeda
perbedaan_kategori = hasil_prediksi[hasil_prediksi['Category Match'] == False]

# Tampilkan hasil
display(perbedaan_kategori)
```
